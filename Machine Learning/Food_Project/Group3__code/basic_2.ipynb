```python
# ‚ñë‚ñë 1. Mount Google Drive ‚ñë‚ñë
from google.colab import drive

# Mount drive
drive.mount('/content/drive')

# ‚ñë‚ñë 2. Imports ‚ñë‚ñë
import os, time, numpy as np
from collections import Counter
from PIL import Image, UnidentifiedImageError
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Dataset, WeightedRandomSampler
from torchvision import transforms, datasets
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    confusion_matrix, roc_auc_score, roc_curve
)
from sklearn.preprocessing import label_binarize

# ‚ñë‚ñë 3. CombinedDataset ‚ñë‚ñë
class CombinedDataset(Dataset):
    def __init__(self, roots, transform=None, extensions=datasets.folder.IMG_EXTENSIONS):
        self.roots = roots
        self.transform = transform
        self.extensions = extensions
        self.classes, self.class_to_idx = self._find_classes()
        self.samples = self._gather_samples()
        print(f"üì¶ Found {len(self.samples)} images across {len(self.classes)} classes.")

    def _find_classes(self):
        all_classes = set()
        for root in self.roots:
            all_classes |= {d.name for d in os.scandir(root) if d.is_dir()}
        classes = sorted(all_classes)
        return classes, {c: i for i, c in enumerate(classes)}

    def _gather_samples(self):
        allowed = datasets.folder.has_file_allowed_extension
        samples = []
        for root in self.roots:
            for cls in self.classes:
                d = os.path.join(root, cls)
                if not os.path.isdir(d): continue
                for fname in os.listdir(d):
                    path = os.path.join(d, fname)
                    if allowed(path, self.extensions):
                        samples.append((path, self.class_to_idx[cls]))
        return samples

    def __len__(self): return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        try:
            img = Image.open(path).convert('RGB')
        except (FileNotFoundError, UnidentifiedImageError):
            return self.__getitem__((idx + 1) % len(self))
        if self.transform:
            img = self.transform(img)
        return img, label

# ‚ñë‚ñë 4. Helpers ‚ñë‚ñë
def count_labels(dataset, class_names, name):
    counts = Counter(y for _, y in dataset)
    print(f"üìä {name} split:")
    for i, cls in enumerate(class_names):
        print(f"   {cls:>15}: {counts.get(i, 0)}")
    print()

def compute_weights(dataset, num_classes):
    label_counts = Counter(y for _, y in dataset)
    weights = [1.0 / label_counts[y] for _, y in dataset]
    return weights

def load_dataset(roots, img_size=224, batch=32, split=0.8):
    train_tfm = transforms.Compose([
        transforms.Resize((img_size, img_size)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
        transforms.ToTensor(),
        transforms.Normalize([0.5] * 3, [0.5] * 3)
    ])
    val_tfm = transforms.Compose([
        transforms.Resize((img_size, img_size)),
        transforms.ToTensor(),
        transforms.Normalize([0.5] * 3, [0.5] * 3)
    ])

    full = CombinedDataset(roots, transform=None)
    classes = full.classes

    n_train = int(len(full) * split)
    n_val = len(full) - n_train
    train_set, val_set = random_split(full, [n_train, n_val])

    train_set.dataset.transform = train_tfm
    val_set.dataset.transform = val_tfm

    count_labels(train_set, classes, "Train")
    count_labels(val_set, classes, "Val")

    weights = compute_weights(train_set, len(classes))
    sampler = WeightedRandomSampler(weights, num_samples=len(weights), replacement=True)

    train_loader = DataLoader(train_set, batch_size=batch, sampler=sampler, num_workers=2)
    val_loader = DataLoader(val_set, batch_size=batch, shuffle=False, num_workers=2)
    return train_loader, val_loader, classes

# ‚ñë‚ñë 5. SimpleCNN with BatchNorm + Dropout ‚ñë‚ñë
class SimpleCNN(nn.Module):
    def __init__(self, n_cls):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1), nn.BatchNorm2d(32), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(), nn.MaxPool2d(2),
            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(128 * 28 * 28, 256), nn.ReLU(),
            nn.Linear(256, n_cls)
        )

    def forward(self, x): return self.net(x)

# ‚ñë‚ñë 6. Train Function (Early Stop + Scheduler) ‚ñë‚ñë
def train(model, train_loader, val_loader, device, epochs=100, lr=1e-4, save='model.pth', patience=10):
    crit = nn.CrossEntropyLoss()
    opt = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    sched = optim.lr_scheduler.StepLR(opt, step_size=20, gamma=0.5)

    best_val = -1
    stop_counter = 0
    tr_hist, vl_hist = [], []

    for ep in range(1, epochs + 1):
        model.train()
        total, correct, loss_sum = 0, 0, 0
        for x, y in tqdm(train_loader, desc=f"Epoch {ep}/{epochs}"):
            x, y = x.to(device), y.to(device)
            opt.zero_grad()
            out = model(x)
            loss = crit(out, y)
            loss.backward()
            opt.step()
            total += y.size(0)
            correct += (out.argmax(1) == y).sum().item()
            loss_sum += loss.item() * x.size(0)

        tr_acc = 100 * correct / total
        model.eval()
        val_acc = val_accuracy(model, val_loader, device)

        tr_hist.append(tr_acc)
        vl_hist.append(val_acc)
        sched.step()

        print(f"  Loss {loss_sum / total:.4f} | Train {tr_acc:.2f}% | Val {val_acc:.2f}%")

        if val_acc > best_val:
            best_val = val_acc
            stop_counter = 0
            torch.save(model.state_dict(), save)
            print(f"  üìÖ saved best model at epoch {ep}")
        else:
            stop_counter += 1
            if stop_counter >= patience:
                print("‚ùå Early stopping triggered.")
                break

    plt.plot(tr_hist, label='Train'); plt.plot(vl_hist, label='Val')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy (%)'); plt.title('Accuracy'); plt.legend(); plt.grid(); plt.show()

@torch.no_grad()
def val_accuracy(model, loader, device):
    model.eval()
    correct = total = 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        out = model(x)
        correct += (out.argmax(1) == y).sum().item()
        total += y.size(0)
    return 100 * correct / total

# ‚ñë‚ñë 7. Report ‚ñë‚ñë
@torch.no_grad()
def report(model, loader, classes, device):
    model.eval()
    y_true, y_pred, y_prob = [], [], []
    for x, y in loader:
        out = model(x.to(device))
        y_true.append(y.numpy())
        y_pred.append(out.argmax(1).cpu().numpy())
        y_prob.append(torch.softmax(out, 1).cpu().numpy())

    y = np.concatenate(y_true)
    p = np.concatenate(y_pred)
    prob = np.vstack(y_prob)

    print(f"\nüìä Accuracy: {accuracy_score(y,p):.4f}")
    print(f"Precision: {precision_score(y,p,average='weighted'):.4f}")
    print(f"Recall:    {recall_score(y,p,average='weighted'):.4f}\n")
    print("Confusion matrix:\n", confusion_matrix(y,p))

    try:
        y_bin = label_binarize(y, classes=range(len(classes)))
        auc = roc_auc_score(y_bin, prob, average='weighted', multi_class='ovr')
        print(f"ROC-AUC: {auc:.4f}")
        for i, cls in enumerate(classes):
            fpr, tpr, _ = roc_curve((y==i).astype(int), prob[:, i])
            plt.plot(fpr, tpr, label=f"{cls} (AUC={roc_auc_score((y==i), prob[:,i]):.2f})")
        plt.plot([0,1],[0,1],'k--'); plt.xlabel('FPR'); plt.ylabel('TPR')
        plt.title('ROC Curves'); plt.legend(); plt.grid(); plt.show()
    except Exception as e:
        print("ROC-AUC not available:", e)

# ‚ñë‚ñë 8. Predict One ‚ñë‚ñë
def predict_one(model, img_path, classes, device):
    tfm = transforms.Compose([
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize([0.5]*3,[0.5]*3)
    ])
    img = Image.open(img_path).convert('RGB')
    pred = model(tfm(img).unsqueeze(0).to(device)).argmax(1).item()
    print(f"\nüîÆ {os.path.basename(img_path)} ‚ûû {classes[pred]}")

# ‚ñë‚ñë 9. Main ‚ñë‚ñë
def main():
    roots = [
        '/content/drive/MyDrive/Dataset/Dataset/Our dataset',
        '/content/drive/MyDrive/Dataset/Dataset/Existing dataset'
    ]
    save_path = '/content/drive/MyDrive/Dataset/Dataset/sami_food_model_final.pth'
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print("‚öôÔ∏è Device:", device)

    train_loader, val_loader, classes = load_dataset(roots)

    model = SimpleCNN(len(classes)).to(device)
    train(model, train_loader, val_loader, device, epochs=100, lr=1e-5, save=save_path)

    model.load_state_dict(torch.load(save_path, map_location=device))
    report(model, val_loader, classes, device)

    predict_one(model, f"{roots[0]}/Mango pudding/IMG20250608133835.jpg", classes, device)

if __name__ == '__main__':
    main()
```
